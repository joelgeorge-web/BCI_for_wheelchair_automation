```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Signal from CSV
df = pd.read_csv(r"D:\Joel\Bioamp_Data\Data\SSVEP\23 Jan\Transposed\10.5_30s_1.csv")

# Assuming the signal is in the first row, you can extract it
signal = df.iloc[0].values

# Step 2: Apply FFT
fft_result = np.fft.fft(signal)

# Step 3: Calculate the Frequency Bins
sampling_rate = 10000  # Adjust if your signal has a different sampling rate
frequencies = np.fft.fftfreq(len(fft_result), d=1/sampling_rate)

# Step 4: Plot the Results
plt.figure(figsize=(10, 6))
# Set x-axis ticks to multiples of 10
plt.xticks(np.arange(0, max(frequencies), 10))

# Set x-axis limits
plt.xlim(0, max(frequencies))

# Plot the magnitude spectrum
plt.subplot(2, 1, 1)
plt.plot(frequencies, np.abs(fft_result))
plt.title('FFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# Plot the phase spectrum
plt.subplot(2, 1, 2)
plt.plot(frequencies, np.angle(fft_result))
plt.title('FFT Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Signal from CSV
df = pd.read_csv(r"D:\Joel\Bioamp_Data\Data\SSVEP\23 Jan\Transposed\Joel_OOR_6_1min_1.csv")

# Assuming the signal is in the first row, you can extract it
signal = df.iloc[0].values

# Step 2: Apply FFT
fft_result = np.fft.fft(signal)

# Step 3: Calculate the Frequency Bins
sampling_rate = 10000  # Adjust if your signal has a different sampling rate
frequencies = np.fft.fftfreq(len(fft_result), d=1/sampling_rate)

# Find the index corresponding to the maximum amplitude
max_amplitude_index = np.argmax(np.abs(fft_result))

# Extract the frequency value with maximum amplitude
freq_with_max_amplitude = frequencies[max_amplitude_index]

# Step 4: Plot the Results
plt.figure(figsize=(10, 6))

# Plot the magnitude spectrum
plt.subplot(2, 1, 1)
plt.plot(frequencies, np.abs(fft_result))
plt.title('FFT Magnitude Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Magnitude')
plt.grid(True)

# Set x-axis ticks to multiples of 10
plt.xticks(np.arange(0, max(frequencies), 10))

# Set x-axis limits
plt.xlim(0, max(frequencies))

# Highlight the frequency with maximum amplitude
plt.scatter(freq_with_max_amplitude, np.abs(fft_result[max_amplitude_index]), color='red', label=f'Max Amp at {freq_with_max_amplitude:.2f} Hz')
plt.legend()

# Plot the phase spectrum
plt.subplot(2, 1, 2)
plt.plot(frequencies, np.angle(fft_result))
plt.title('FFT Phase Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase')
plt.grid(True)

# Set x-axis ticks to multiples of 10
plt.xticks(np.arange(0, max(frequencies), 10))

# Set x-axis limits
plt.xlim(0, max(frequencies))

plt.tight_layout()
plt.show()

# Print the frequency value with maximum amplitude
print(f"The frequency with maximum amplitude is {freq_with_max_amplitude:.2f} Hz.")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Step 1: Load the Signal from CSV
df = pd.read_csv(r"D:\Joel\Bioamp_Data\Data\SSVEP\23 Jan\Transposed\Joel_OOR_6_1min_1.csv")

# Step 2: Get the number of rows and columns in the DataFrame
num_rows, num_columns = df.shape

# Step 3: Set the sampling rate
sampling_rate = 10000  # Adjust if your signal has a different sampling rate

# Iterate through each row
for row_index in range(num_rows):
    # Extract the signal from the current row
    signal = df.iloc[row_index].values

    # Apply FFT
    fft_result = np.fft.fft(signal)

    # Calculate the Frequency Bins
    frequencies = np.fft.fftfreq(len(fft_result), d=1/sampling_rate)

    # Find the index corresponding to the maximum amplitude
    max_amplitude_index = np.argmax(np.abs(fft_result))

    # Extract the frequency value with maximum amplitude
    freq_with_max_amplitude = frequencies[max_amplitude_index]

    # Plot the Results
    plt.figure(figsize=(10, 6))

    # Plot the magnitude spectrum
    plt.subplot(2, 1, 1)
    plt.plot(frequencies, np.abs(fft_result))
    plt.title(f'FFT Magnitude Spectrum - Row {row_index + 1}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Magnitude')
    plt.grid(True)

    # Set x-axis ticks to multiples of 10
    plt.xticks(np.arange(0, max(frequencies), 10))

    # Set x-axis limits
    plt.xlim(0, max(frequencies))

    # Highlight the frequency with maximum amplitude
    plt.scatter(freq_with_max_amplitude, np.abs(fft_result[max_amplitude_index]), color='red', label=f'Max Amp at {freq_with_max_amplitude:.2f} Hz')
    plt.legend()

    # Plot the phase spectrum
    plt.subplot(2, 1, 2)
    plt.plot(frequencies, np.angle(fft_result))
    plt.title(f'FFT Phase Spectrum - Row {row_index + 1}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Phase')
    plt.grid(True)

    # Set x-axis ticks to multiples of 10
    plt.xticks(np.arange(0, max(frequencies), 10))

    # Set x-axis limits
    plt.xlim(0, max(frequencies))

    plt.tight_layout()
    plt.show()

    # Print the frequency value with maximum amplitude for each row
    print(f"For Row {row_index + 1}, the frequency with maximum amplitude is {freq_with_max_amplitude:.2f} Hz.")
```

```python
import pandas as pd
import numpy as np
from scipy.fft import fft
import matplotlib.pyplot as plt

# Function to calculate FFT and find frequency of max amplitude
def calculate_fft(chunk, sampling_rate):
    signal1 = chunk['Signal1'].values
    signal2 = chunk['Signal2'].values

    # FFT for each signal
    fft_signal1 = fft(signal1)
    fft_signal2 = fft(signal2)

    # Combined FFT
    combined_fft = fft_signal1 + fft_signal2

    # Frequency calculation
    frequencies = np.fft.fftfreq(len(chunk), 1/sampling_rate)
    max_freq_index = np.argmax(np.abs(combined_fft))
    max_freq = frequencies[max_freq_index]

    return frequencies, fft_signal1, fft_signal2, combined_fft, max_freq

# Read CSV file in chunks of 25000 rows
chunk_size = 25000
sampling_rate = 5000  # Adjust the sampling rate accordingly
file_path = r"D:\Joel\Bioamp_Data\Data\SSVEP\23 Jan\Joel_OOR_6_1min_1.csv"  # Replace with the actual file path

# Iterate through chunks
for chunk in pd.read_csv(file_path, chunksize=chunk_size):
    frequencies, fft_signal1, fft_signal2, combined_fft, max_freq = calculate_fft(chunk, sampling_rate)

    # Plot the FFT for each signal and combined
    plt.figure(figsize=(10, 6))
    plt.plot(frequencies, np.abs(fft_signal1), label='Signal 1 FFT')
    plt.plot(frequencies, np.abs(fft_signal2), label='Signal 2 FFT')
    plt.plot(frequencies, np.abs(combined_fft), label='Combined FFT')
    plt.title('FFT Analysis')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.legend()
    plt.show()

    print(f"Frequency of Max Amplitude: {max_freq} Hz")
```

```python
import pandas as pd
import numpy as np
from scipy.fft import fft
import matplotlib.pyplot as plt
import seaborn as sns

# Function to calculate FFT and find frequency of max amplitude
def calculate_fft(chunk, sampling_rate):
    signal1 = chunk.iloc[:, 0].values  # Assuming Signal 1 is in the first column
    signal2 = chunk.iloc[:, 1].values  # Assuming Signal 2 is in the second column

    # FFT for each signal
    fft_signal1 = fft(signal1)
    fft_signal2 = fft(signal2)

    # Combined FFT
    combined_fft = fft_signal1 + fft_signal2

    # Frequency calculation
    frequencies = np.fft.fftfreq(len(chunk), 1/sampling_rate)
    max_freq_index = np.argmax(np.abs(combined_fft))
    max_freq = frequencies[max_freq_index]

    return frequencies, fft_signal1, fft_signal2, combined_fft, max_freq

# Read CSV file in chunks of 25000 rows
chunk_size = 25000
sampling_rate = 5000  # Adjust the sampling rate accordingly
file_path = r"D:\Joel\Bioamp_Data\Data\SSVEP\23 Jan\Joel_OOR_19_1min_1.csv"  # Replace with the actual file path

# Initialize a list to store max frequencies
max_freq_list = []

# Iterate through chunks
for i, chunk in enumerate(pd.read_csv(file_path, header=None, chunksize=chunk_size)):
    frequencies, fft_signal1, fft_signal2, combined_fft, max_freq = calculate_fft(chunk, sampling_rate)
    max_freq_list.append(max_freq)

    # Plot the FFT for each signal and combined
    plt.figure(figsize=(12, 8))
    plt.subplot(2, 1, 1)
    plt.plot(frequencies, np.abs(fft_signal1), label='Signal 1 FFT')
    plt.plot(frequencies, np.abs(fft_signal2), label='Signal 2 FFT')
    plt.plot(frequencies, np.abs(combined_fft), label='Combined FFT')
    plt.title(f'FFT Analysis - Chunk {i+1}')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Amplitude')
    plt.legend()

    # Plot the FFT Heatmap
    plt.subplot(2, 1, 2)
    sns.heatmap(np.abs(combined_fft).reshape(1, -1), cmap='viridis', xticklabels=frequencies)
    plt.title('FFT Heatmap - Combined FFT')
    plt.xlabel('Frequency (Hz)')
    plt.ylabel('Chunk Index')

    plt.tight_layout()
    plt.show()

# Plot the frequency distribution
plt.figure(figsize=(10, 6))
plt.hist(max_freq_list, bins=20, edgecolor='black')
plt.title('Frequency Distribution of Max Amplitude')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Number of Chunks')
plt.show()

# Print the overall max frequency
overall_max_freq = max(max_freq_list)
print(f"Overall Max Frequency: {overall_max_freq} Hz")
```
